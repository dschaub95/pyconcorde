```python
import os
atoi = lambda text : int(text) if text.isdigit() else text
natural_keys = lambda text : [atoi(c) for c in re.split('(\d+)', text)]
fnames = os.listdir(data_dir)
fnames.sort(key=natural_keys)
lengths = []
solutions = []
for fname in fnames:
    try:
        solver = TSPSolver.from_tspfile(data_dir + fname)
        solution = solver.solve()
        lengths.append(solution.optimal_value/(10000)) # *solution.tour.shape[0]))
        solutions.append(solution.tour)
    except:
        print("Could not load file {}".format(fname))
final_solution = np.sum(lengths)/np.array(lengths).shape[0]
print(final_solution)
```

```python
fnames = os.listdir(data_dir)
fnames.sort(key=natural_keys)
path = data_dir
file_name = 'lengths.txt'
with open(path+file_name, 'w') as f:
    for i, fname in enumerate(fnames):
        if 'tsp' in fname:
            f.write('{}: {}\n'.format(fname, lengths[i]))
        else:
            continue
file_name = 'solutions.txt'
with open(path+file_name, 'w') as f:
    for i, fname in enumerate(fnames):
        if 'tsp' in fname:
            f.write('{}: {}\n'.format(fname, list(solutions[i])))
        else:
            continue
```

```python
with open(path+'lengths.txt', 'r') as f:
    lines = f.readlines()
    lines = [float(line.split(':')[-1].strip()) for line in lines]
lines
```

```python
index = pd.RangeIndex(start=10, stop=20)
sol_df = pd.DataFrame()
for k, fname in enumerate(fnames):
    tmp_df = pd.DataFrame()
    tmp_df[fname.split('.')[0]] = solutions[k]
    sol_df = pd.concat([sol_df,tmp_df.astype(int)], ignore_index=False, axis=1)
```

```python
sol_df.to_csv('results/{}/solution_{}.csv'.format(data_dir.split("/")[-1], 'concorde'))
```

```python
lens_df = pd.DataFrame()
for k, fname in enumerate(fnames):
    tmp_df = pd.DataFrame()
    tmp_df[fname.split('.')[0]] = [lengths[k]]
    lens_df = pd.concat([lens_df,tmp_df], ignore_index=False, axis=1)
```
